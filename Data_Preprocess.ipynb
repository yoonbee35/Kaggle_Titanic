```python
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Preprocessing
### 데이터 분석을 통해 얻은 정보
1. *Ticket*(생존 여부와 관계 없음), *Name, PassengerId*(모두 다른 값을 가지고 있고 생존 여부과 관계 없음), *Cabin*(train, test데이터 모두 너무 많은 Null값을 가지고 있음) 등의 데이터는 feature로 사용하지 않음
2. Age와 Embarked는 생존여부와 관계가 있다고 생각되기 때문에 Null값을 채워넣어야 함
3. Age와 Fare(연속형 데이터)을 범주형 데이터로 만들어 일반화
4. 여성이 남성보다 많이 생존했기 때문에 Sex는 유의미한 feature ==> numeric data로 변환

** feature들 사이의 scale 차이가 많이 남(0 ~ 700) --> 특성 scaling 진행 必 <br>

## Data로 사용하지 않은 특성 제거하기 : Name, PassengerId, Ticket, Cabin

```python
print("Before :", train.shape, test.shape)

train_df = train.drop(['Ticket', 'Cabin', 'Name', 'PassengerId'], axis=1)
test_df = test.drop(['Ticket', 'Cabin', 'Name'], axis=1) #test의 'PassengerId'는 제출파일에 포함되어야하므로 제거 안함
combine = [train_df, test_df]

print("After :", train_df.shape, test_df.shape)
```

```python
# Sex data를 numeric으로 변환하기
# female=1, male=0

for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)

train_df.head()
```

## 특성들을 조합하여 새로운 feature 만들기

```python
# 형제자매 + 자녀 => Family라는 feature 생성
for dataset in combine:
    dataset['Family'] = dataset['SibSp'] + dataset['Parch'] + 1

train_df[['Family', 'Survived']].groupby(['Family'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
# SibSp와 Parch라는 feature을 없애고 Family로 대체
train_df = train_df.drop(['SibSp', 'Parch'], axis=1)
test_df = test_df.drop(['SibSp', 'Parch'], axis=1)

combine = [train_df, test_df]
```

```python
train_df.head(10)
```

## Null값 채우기
### Age

```python
# Pclass와 Sex를 이용하여 Age 예측 ==> 그 예측한 중앙값으로 Null값 채우기

grid = sns.FacetGrid(train_df, row='Pclass', col='Sex', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python
guess_ages = np.zeros((2,3))
guess_ages

for dataset in combine:
    for i in range(0, 2):
        for j in range(0, 3):
            guess_df = dataset[(dataset['Sex'] == i) & 
                                  (dataset['Pclass'] == j+1)]['Age'].dropna()

            # age_mean = guess_df.mean()
            # age_std = guess_df.std()
            # age_guess = rnd.uniform(age_mean - age_std, age_mean + age_std)

            age_guess = guess_df.median() # 중앙값 사용

            # Convert random age float to nearest .5 age
            guess_ages[i,j] = int( age_guess/0.5 + 0.5 ) * 0.5
            
    for i in range(0, 2):
        for j in range(0, 3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),\
                    'Age'] = guess_ages[i,j]

    dataset['Age'] = dataset['Age'].astype(int)

train_df.head()
```

```python
train_df['AgeBand'] = pd.cut(train_df['Age'], 5)
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
# Age를 범주형 데이터로 변환
for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4

# 0~16세=0, 17~32세=1, 33~48세=2, 49~64세=3, 65세~ =4
train_df = train_df.drop(['AgeBand'], axis=1)
combine = [train_df, test_df]
train_df.head()
```

### Embarked *최빈값으로 채우기

```python
freq_port = train_df.Embarked.dropna().mode()[0]
freq_port
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)
    
train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
# object data => numeric data
# S=0, C=1, Q=2
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

train_df.head()
```

### Fare (test data에만 Null값 존재)

```python
# Fare도 Null값 채운 후 범주형 데이터로 변환
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True)
test_df.head()
```

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
# 돈을 많이 낸 승객일수록 생존 확률 높음
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train_df = train_df.drop(['FareBand'], axis=1)
combine = [train_df, test_df]
    
train_df.head(10)
```

```python
# 다른 요소가 추가되어 영향을 받지 않도록 index=False로 표시

train_df.to_csv("train_process.csv",index=False)
test_df.to_csv("test_process.csv",index=False)
```
